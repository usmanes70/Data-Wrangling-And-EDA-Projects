# Price Per Liter (PKR) in each country (Interactive Visualization)

```python
fig = px.bar(pg_price, y='Price Per Liter (PKR)', x='Country', text_auto='.2s', color = 'Country',
            title="Price Per Liter (PKR) in each country",
            hover_data = ['GDP Per Capita ( USD )', 'Gallons GDP Per Capita Can Buy', 'xTimes Yearly Gallons Per Capita Buy'],)

fig.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)
fig.update_layout(
    height = 800,
    width=1600
)
fig.show()
```

# Top 10 Countries in Which Petrol Prices are High

```python
# Find the largest values in price per liter column
top10_high = pg_price.nlargest(10, 'Price Per Liter (PKR)')
top10_high
```

## Interactive visualization of top 10 countries

```python
fig = px.bar(top10_high, x='Country', y= 'Price Per Liter (USD)', 
             hover_data = ['GDP Per Capita ( USD )', 'Gallons GDP Per Capita Can Buy', 'xTimes Yearly Gallons Per Capita Buy'],  
             color = 'Country')

fig.show()
```

**The top 10 countries where petrol prices are high are north korea, tonga, niue, hongkong, norway, denmark, finland, iceland, greece and netherlands.** 

**Those countries where petrol prices are high gdp per capita is less and where petrol prices are less, gdp per capita is greater.**

# Comparison of fuel prices in Pakistan with US, China, India, and Bangladesh 

```python
# Make a subset of countries
pak = pg_price[pg_price['Country'] == 'Pakistan']
us = pg_price[pg_price['Country'] == 'United States']
china = pg_price[pg_price['Country'] == 'China']
india = pg_price[pg_price['Country'] == 'India']
bangladesh = pg_price[pg_price['Country'] == 'Bangladesh']

# Concatenate to make a dataframe to compare the countries
df_concat = pd.concat([pak, us, china, india, bangladesh])

df_concat
```

```python
fig = px.bar(df_concat, y='Price Per Liter (PKR)', x='Country', text_auto='.2s', color = 'Country',
            title="Comparison of fuel prices",
            hover_data = ['GDP Per Capita ( USD )', 'Gallons GDP Per Capita Can Buy', 'xTimes Yearly Gallons Per Capita Buy'])

fig.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)

fig.show()
```

# Visualize Daily Oil Consumptions Of above Countries

```python
fig = px.bar(df_concat, y='Daily Oil Consumption (Barrels)', x='Country', color = 'Country', 
            title="Daily Oil Consumption In Barrels",
            hover_data = ['GDP Per Capita ( USD )', 'Gallons GDP Per Capita Can Buy', 'xTimes Yearly Gallons Per Capita Buy'],
            log_y = True)

fig.show()
```

# Correlation between the variables

### Typecasting

We do typecasting to convert the object type into int64 type and also replace the comma in the values with the space.

```python
pg_price['Daily Oil Consumption (Barrels)'] = pg_price['Daily Oil Consumption (Barrels)'].str.replace(',', '').astype('int64')
```

```python
pg_price['GDP Per Capita ( USD )'] = pg_price['GDP Per Capita ( USD )'].str.replace(',', '').astype('int64')
```

```python
pg_price['Gallons GDP Per Capita Can Buy'] = pg_price['Gallons GDP Per Capita Can Buy'].str.replace(',', '').astype('int64')
```

```python
# Drop the country variable
new_df = pg_price.drop('Country', axis = 1)
```

```python
# Check correlation
new_df.corr()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(new_df.corr(), annot=True, fmt='.2f')
```

**Positive Correlation exists between the variables:**

1. Daily Oil Consumption (Barrels) and Yearly Gallons Per Capita
2. Daily Oil Consumption (Barrels) and GDP Per Capita ( USD )
3. Yearly Gallons Per Capita and Daily Oil Consumption (Barrels)
4. Yearly Gallons Per Capita and GDP Per Capita ( USD )
5. Price Per Liter (USD) and GDP Per Capita ( USD )
6. xTimes Yearly Gallons Per Capita Buy and Gallons GDP Per Capita Can Buy

**It means on increase of one variable, other variable will be increases or vice-versa.**
